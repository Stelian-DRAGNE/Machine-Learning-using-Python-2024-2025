```python
from scipy import stats
```

```python
rezultat = stats.linregress(suprafete, preturi)
rezultat
```

```python
def f_perfect(x):
    a = rezultat.slope
    b = rezultat.intercept
    return a * x + b
preturi_generate_cu_f_perfect = np.array(list(map(f_perfect, suprafete)))
preturi_generate_cu_f_perfect
plt.plot(suprafete, preturi_generate_cu_f_perfect, color="red")

def f1(x):
    a = 2.2
    b = 5
    return a * x + b
preturi_generate_cu_f_1 = np.array(list(map(f1, suprafete)))
preturi_generate_cu_f_1
plt.plot(suprafete, preturi_generate_cu_f_1, color="blue")


def f2(x):
    a = 1.9
    b = -2
    return a * x + b
preturi_generate_cu_f_2 = np.array(list(map(f2, suprafete)))
preturi_generate_cu_f_2
plt.plot(suprafete, preturi_generate_cu_f_2, color="purple")

plt.plot(suprafete, preturi, color="green")
plt.scatter(suprafete, preturi, color="red")

plt.xlabel("Suprafa (m)")
plt.ylabel("Pre (mii )");
```
