```python
import numpy as np
```

```python
x = np.arange(5)
y = np.arange(5) * 2

plt.scatter(x, y);
```

```python
x = np.arange(5)
y = np.arange(5) * 2

plt.plot(x, y);
```

```python
x = np.arange(5)
y = np.arange(5) * 2

plt.scatter(x, y);
plt.plot(x, y);
```

```python
x = np.arange(5)
y = np.arange(5) * 2

plt.scatter(x, y);
plt.plot(x, y);
plt.bar(x, y);
```

```python
x = np.arange(5)
y = np.arange(5) * 2

plt.scatter(x, y, color="black");
plt.plot(x, y);
plt.bar(x, y, color = "purple");
```

```python
x = np.linspace(2, 8, 5)
y = np.linspace(2, 8, 5) * 2

plt.scatter(x, y, color="black");
plt.plot(x, y);
plt.bar(x, y, color = "purple");
```

```python
x = np.random.randint(2, 8, 5)
y = np.random.randint(2, 8, 5) * 2

plt.scatter(x, y, color="black");
plt.plot(x, y);
```
