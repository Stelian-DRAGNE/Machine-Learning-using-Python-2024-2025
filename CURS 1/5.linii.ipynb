```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    a = 0
    b = 0
    return a * x + b

suprafete = np.arange(40, 51)
suprafete
```

```python
preturi = []
for s in suprafete:
    preturi.append(f(s))
print(preturi)
```

```python
def f(x):
    a = 0
    b = 0
    return a * x + b

suprafete = np.arange(40, 51)

preturi = np.array( list(map(f, suprafete)) )

plt.scatter(suprafete, preturi, color="red")
plt.plot(suprafete, preturi, color="blue");
```
