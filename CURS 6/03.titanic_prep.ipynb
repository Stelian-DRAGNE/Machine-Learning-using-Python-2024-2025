```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv("titanic/train.csv")
test_df = pd.read_csv("titanic/test.csv")
```

```python
train_df.head(2)
```

```python
test_df.head(2)
```

```python
train_survived = train_df["Survived"]
```

```python
train_df = train_df.drop(columns=["Survived"], axis=1)
```

```python
train_df.head(1)
```

```python
train_df = train_df.drop(["Ticket", "Name"], axis=1)
test_df = test_df.drop(["Ticket", "Name"], axis=1)
```

```python
train_df.head(1)
```

```python
train_df["Sex"] = train_df["Sex"] == "male"
test_df["Sex"] = test_df["Sex"] == "male"
train_df.head(1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
train_df["Embarked"] = LabelEncoder().fit_transform(train_df["Embarked"])
test_df["Embarked"] = LabelEncoder().fit_transform(test_df["Embarked"])
train_df.head(1)
```

```python
train_df["Cabin"] = train_df["Cabin"].astype(str).str[0]
test_df["Cabin"] = test_df["Cabin"].astype(str).str[0]
train_df.head(1)
```

```python
train_df["Cabin"] = LabelEncoder().fit_transform(train_df["Cabin"])
test_df["Cabin"] = LabelEncoder().fit_transform(test_df["Cabin"])
train_df.head(1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
train_df.keys()
```

```python
train_df = StandardScaler().fit_transform(train_df)
test_df = StandardScaler().fit_transform(test_df)
```

```python
train_df = pd.DataFrame(data=train_df, columns = ['PassengerId', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin', 'Embarked'])
test_df = pd.DataFrame(data=test_df, columns = ['PassengerId', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin', 'Embarked'])
```

```python
train_df.head(2)
```

```python
test_df.head(2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(max_depth=3)
model
```

```python
model.fit(train_df, train_survived)
```

```python
test_survived = pd.read_csv("titanic/gender_submission.csv", index_col="PassengerId")
test_survived
```

```python
model.score(test_df, test_survived)
```

```python
from sklearn.tree import export_graphviz
```

```python
exported_data = export_graphviz(model, feature_names=['PassengerId', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare',
    'Cabin', 'Embarked'], class_names=["Not Survived", "Survived"], filled=True, rounded=True)
```

```python
exported_data
```

```python
import graphviz
```

```python
graph = graphviz.Source(exported_data)
graph.render("Titanic Tree structure", format="png")
graph.view()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(train_df, train_survived)
```

```python
rf_model.estimator
```

```python
found_estimator = rf_model.estimators_[0]
found_estimator.fit(train_df, train_survived)
```

```python
exported_data
```
