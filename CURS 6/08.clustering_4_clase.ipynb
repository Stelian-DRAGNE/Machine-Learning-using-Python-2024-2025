```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("04_categories.csv")
x = df[["coord_x", "coord_y"]]
y = df["category"]
```

```python
plt.scatter(df["coord_x"], df["coord_y"]);
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=7)
```

```python
model.fit(x, y)
```

```python
model.labels_
```

```python
plt.scatter(df["coord_x"], df["coord_y"], c=model.labels_);
```

```python
d_colors = { "a" : "red", "b" : "blue", "c" : "green", "d" : "yellow" }
plt.scatter(df["coord_x"], df["coord_y"], c=df["category"].replace(d_colors));
```
