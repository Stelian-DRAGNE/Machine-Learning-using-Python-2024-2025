# https://www.kaggle.com/datasets/aasthavrm/titanic-database

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = pd.read_csv("titanic/train.csv")
x_test = pd.read_csv("titanic/test.csv")
```

```python
x_train
```

```python
variabile_independente  = x_train[["Sex", "Age"]]
variabile_independente["Sex"] = variabile_independente["Sex"] == "male"
```

```python
variabila_dependenta  = x_train[["Survived"]]
```

```python
plt.scatter(variabile_independente["Age"], variabila_dependenta, c=variabile_independente["Sex"]);
```

```python
plt.rcParams["figure.figsize"] = [12.8, 4.8]
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(variabile_independente, variabila_dependenta)
```

```python
variabile_de_test_independente  = x_test[["Sex", "Age"]]
variabile_de_test_independente["Sex"] = variabile_de_test_independente["Sex"] == "male"
```

```python
variabile_de_test_independente
```

```python
variabile_de_test_dependenta = pd.read_csv("titanic/gender_submission.csv")
```

```python
variabile_de_test_dependenta = variabile_de_test_dependenta["Survived"]
```

```python
rf_model.score(variabile_de_test_independente, variabile_de_test_dependenta)
```
