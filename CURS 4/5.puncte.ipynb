```python
import pandas as pd
```

```python
df = pd.read_csv("puncte.csv")
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df["coord_x"], df["coord_y"], c=df["categoria"]=="a");
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
model
```

```python
model.fit(df[["coord_x", "coord_y"]], df["categoria"])
```

```python
model.predict([[100, 200]])
```
